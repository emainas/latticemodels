```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bJ = 0.01

def gamma(x):
    return np.log( (x*np.exp(-2*bJ) + np.sqrt(1 - x*x*(1 - np.exp(-4*bJ))) )/(np.sqrt(1-x*x)))
```

```python
mu = np.arange(-0.5,0.501,0.01)
plt.plot(mu, gamma(mu), lw=3, color='blue')
plt.plot(mu, 0*mu, lw=2, color='black')
plt.xlim(-0.5, 0.5)
plt.xlabel(r'$\mu$', fontsize='25')
plt.ylabel(r'$\gamma(\mu)$', fontsize='25')
plt.tick_params(labelsize=20)
plt.title('Equation of State', fontsize= 15, fontweight='bold')
plt.show()
```

```python
a1 = np.exp(2*bJ)
a2 = np.exp(-2*bJ)
a3 = np.exp(-bJ)

def lam(x):
    return np.log((np.sqrt(a1*(1-x*x) + a2*x*x) + a3)  /  (2*np.cosh(bJ)*np.sqrt(1-x*x)))
```

```python
plt.plot(mu, lam(mu), lw=3, color='blue')
plt.plot(mu, 0*mu, lw=2, color='black')
plt.xlim(-0.5, 0.5)
plt.xlabel(r'$\mu$', fontsize='25')
plt.ylabel(r'$\lambda(\mu)$', fontsize='25')
plt.tick_params(labelsize=20)
plt.title('Cumulant Generating Function', fontsize= 15, fontweight='bold')
plt.show()
```

```python
def rate(x):
    return x*gamma(x) - lam(x)
```

```python
plt.plot(mu, rate(mu), lw=3, color='blue')
plt.plot(mu, 0*mu, lw=2, color='black')
plt.xlim(-0.5, 0.5)
plt.xlabel(r'$\mu$', fontsize='25')
plt.ylabel(r'$I(\mu)$', fontsize='25')
plt.tick_params(labelsize=20)
plt.title('Rate Function', fontsize= 15, fontweight='bold')
plt.show()
```

```python
def prob(x):
    return np.exp(-N*rate(x))

N = 1000
```

```python
plt.plot(mu, prob(mu), lw=3, color='blue')
```
