```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special as special
```

```python
a = 1/6
C = (1/2)*np.sqrt(np.pi/3)*(1/np.sqrt(a))

def lam(x):
    return np.log(C*np.exp(-a*x)*(1/np.sqrt(x))*special.erfi(np.sqrt(3*a*x)))

def asym(x):
    return 2*a*x -np.log(6*a) - np.log(x)
```

```python
g = np.arange(0,20,0.01)
g2 = np.vectorize(lam)
g3 = np.vectorize(asym)
plt.plot(g, g2(g), lw=3, color='blue', label='exact')
plt.plot(g, g3(g), lw=3, color='red', label='asymptote')
plt.plot(g, 0*g, lw=2, color='black')

plt.xlim(0, 20)
plt.xlabel(r'$\gamma$', fontsize='25')
plt.ylabel(r'$\lambda(\gamma)$', fontsize='25')
plt.tick_params(labelsize=20)
plt.legend(loc='upper left')
plt.show()
```
